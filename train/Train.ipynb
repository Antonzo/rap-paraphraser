```python
import warnings
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')
import gensim 
import logging
import pandas as pd
import re

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

data = pd.read_csv('normal_data.csv', '|', error_bad_lines=False, encoding = 'UTF-8')
```

```python
stop_words = re.split(' +', '''еще	него
а	ж	нее	со
без	же	ней	совсем
более	нельзя	так
больше	за	нет	такой
будет	зачем	ни	там
будто	здесь	нибудь	тебя
бы	и	никогда	тем
был	из	ним	теперь
была	из-за	них	то
были	или	ничего	тогда
было	им	но	того
быть	ну	тоже
в	их	о	только
вам	к	об	том
вас	кажется	один	тот
вдруг	как	он	три
ведь	какая	она	тут
во	какой	они	ты
вот	когда	опять	у
впрочем	конечно	от	уж
все	которого	перед	уже
всегда	которые	по
всего	кто	под	хоть
всех	куда	после	чего
всю	ли	потом
вы	лучше	потому	чем	между	почти	через
где	меня	при	что	мне	про	чтоб
да	много	раз	чтобы
даже	разве	чуть
два	с	эти
для	мой	сам	этого
до	моя	свое	этой
другой	мы	свою	этом
его	на	себе	этот
ее	над	себя	эту
ей	надо	сегодня	я
ему	наконец	сейчас	
если	нас	
есть	не'''.replace('\t', ' ').replace('\n', ' '))

documents = []
for song in data['text']:
    l = []
    for word in song.split(' '):
        a = word.split('_')
        if a[0] not in stop_words:
            l.append(word)
    documents.append(l)
```

```python
model = gensim.models.Word2Vec (documents, size=200, window=10, min_count=2, workers=10)
model.train(documents,total_examples=len(documents),epochs=10)
```

```python
model.wv.most_similar (positive='лоск_NOUN')
```

```python
model.wv.save_word2vec_format('model.bin', binary=True)
```
