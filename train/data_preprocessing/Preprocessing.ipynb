```python
import pymorphy2
import re
import pandas as pd
import csv 

#Функция для записи данных в CSV 
def csv_writer(data, path): 
  with open(path, "w", newline='') as csv_file: 
  #csv_file - объект с данными 
  #delimiter - разделитель 
    writer = csv.writer(csv_file, delimiter='|') 
    for line in data: 
      if(data == 0): 
        break 
      try: 
        writer.writerow(line) 
      except: 
        pass


m = pymorphy2.MorphAnalyzer()

def my_tag(str):
  a1 = []
  a1 = re.split(' +', str)

  for i,w in enumerate(a1):
    if w == '':
      continue
    bb = m.parse(w)[0].tag.POS
    bb1 = replace(bb)
    a1[i] = w + '_' + bb1
  return a1

def my_tag_normal(str):
  a1 = []
  a1 = re.split(' +', str)
 
  for i,w in enumerate(a1):
    bb0 = m.parse(w)[0].normal_form
    bb = m.parse(bb0)[0].tag.POS
    if(bb is None):
      bb1 = 'X'
    else:
      bb1 = replace(bb)
    print(w)
    a1[i] = bb0 + '_' + bb1
  return a1

def replace(x):
    return {
        'ADJF':  'ADJ',
        'ADJS' : 'ADJ',
        'ADVB' : 'ADV',
        'COMP' : 'ADV',
        'GRND' : 'VERB',
        'INFN' : 'VERB',
        'NOUN' : 'NOUN',
        'PRED' : 'ADV',
        'PRTF' : 'ADJ',
        'PRTS' : 'VERB',
        'VERB' : 'VERB',
        'PREP' : 'PREP',
        'CONJ' : 'CONJ',
        'PRCL' : 'PRCL',
        'INTJ' : 'INTJ',
        'NUMR' : 'NUM',
        'NPRO' : 'PRON'
    }[x]

stop_words = re.split(' +', '''-	еще	него
а	ж	нее	со
без	же	ней	совсем
более	нельзя	так
больше	за	нет	такой
будет	зачем	ни	там
будто	здесь	нибудь	тебя
бы	и	никогда	тем
был	из	ним	теперь
была	из-за	них	то
были	или	ничего	тогда
было	им	но	того
быть	ну	тоже
в	их	о	только
вам	к	об	том
вас	кажется	один	тот
вдруг	как	он	три
ведь	какая	она	тут
во	какой	они	ты
вот	когда	опять	у
впрочем	конечно	от	уж
все	которого	перед	уже
всегда	которые	по
всего	кто	под	хоть
всех	куда	после	чего
всю	ли	потом
вы	лучше	потому	чем	между	почти	через
где	меня	при	что	мне	про	чтоб
да	много	раз	чтобы
даже	разве	чуть
два	с	эти
для	мой	сам	этого
до	моя	свое	этой
другой	мы	свою	этом
его	на	себе	этот
ее	над	себя	эту
ей	надо	сегодня	я
ему	наконец	сейчас	
если	нас	
есть	не'''.replace('\t', ' ').replace('\n', ' '))


data = pd.read_csv('data.csv', '|', error_bad_lines=False, encoding = 'UTF-8')
data.columns = ['author', 'title', 'text']

cnt = len(data['author'])
d = [['']*3]*(cnt + 1)
d[0] = ['author', 'title', 'text']
for i in range(cnt):
  d[i][0] = str(data['author'][i]).replace('\t', ' ').replace('\n', ' ').replace('ё','е')
  d[i][1] = str(data['title'][i]).replace('\t', ' ').replace('\n', ' ').replace('ё','е')
  text = d[i][1].lower() + ' ' + str(data['text'][i]).replace('\t', ' ').replace('\n', ' ').lower().replace('ё','е')

  words = re.split('[ ,\.\!\?:;"\+\'\(\)»]+', text)
  for j in range(len(words)):
    if words[j] == '':
      continue
    if words[j][len(words[j])-1] == '_':
      words[j] = words[j][:len(words[j])-1]
    if (words[j] == '-') or (re.search('[a-z_0-9]', words[j])) or (words[j] in stop_words):
      words[j] = ''
  d[i][3] = my_tag_normal(' '.join(words).strip())
  if i == 3:
    break

if __name__ == "__main__": 
  csv_writer(d, "data_new.csv")
```
